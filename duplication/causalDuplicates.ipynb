```python
import numpy as np
import pandas as pd
import causal_duplicate as cd
import altair as alt
```

```python
X_dim = 10
XT_weights = np.ones(X_dim)/X_dim
XY_weights = XT_weights
XY_T_weights = XT_weights
data = cd.generate_causal_data(1000, X_dim=10, XT_weights=XT_weights, XY_weights=XY_weights, XY_T_weights=XY_T_weights)
data["group"] = np.arange(len(data))
data_dup = cd.duplicate_untreated(data, n_duplicates=20)
data_dup.shape
```

```python
cmat, cmat_dup = cd.correlation_matrix_plot(data), cd.correlation_matrix_plot(data_dup)
cmat | cmat_dup 
```

Joint distribution of $Y$

```python
y_dist = cd.distribution_plot(data, feature_name="Y", density=True)


y_dist
```
